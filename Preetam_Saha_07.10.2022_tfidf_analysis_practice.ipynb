```python
import pandas as pd
```

```python
#https://www.kaggle.com/datasets/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews
```

```python
data = pd.read_csv("IMDB Dataset.csv")
data.head()
```

```python
data.shape
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import cohen_kappa_score
```

```python
data.info()
```

```python
target = {"positive":1,"negative":0}
```

```python
data["sentiment"] = data["sentiment"].map(target)
```

```python
data.head()
```

```python
data.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data["review"],data["sentiment"],test_size=0.3,random_state=123456)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_train[:5]
```

```python
from wordcloud import STOPWORDS
```

```python
tfidf = TfidfVectorizer(token_pattern="[a-z']+",max_df=0.6,min_df=10,stop_words=STOPWORDS)
```

```python
tfidf_mat_train = tfidf.fit_transform(x_train)
tfidf_mat_val = tfidf.transform(x_test)
```

```python
tfidf_mat_train.shape
```

```python
tfidf_mat_train
```

```python
tfidf_mat_val.shape
```

```python
lr = LogisticRegression()
```

```python
lr.fit(tfidf_mat_train,y_train)
```

```python
pred = lr.predict(tfidf_mat_val)
```

```python
pred
```

```python
cohen_kappa_score(pred,y_test)
```
